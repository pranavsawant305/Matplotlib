# subplot: it means multiple plots

```python
plt.figure(figsize=(15,4))
a1=np.arange(1,11)
a2=a1**2
a3=a1**3
a4=a1*2
plt.subplot(1,4,1)
plt.plot(a1,a2)
plt.title("a1 vs a2")
plt.subplot(1,4,2)
plt.plot(a1,a3)
plt.title("a1 vs a3")
plt.subplot(1,4,3)
plt.plot(a1,a4)
plt.title("a1 vs a4")
plt.subplot(1,4,4)
plt.plot(a2,a4)
plt.title("a2 vs a4")

```

```python
plt.figure(figsize=(8,10))
a1=np.arange(1,11)
a2=a1**2
a3=a1**3
a4=a1*2
plt.subplot(4,1,1)
plt.plot(a1,a2)
plt.title("a1 vs a2")
plt.subplot(4,1,2)
plt.plot(a1,a3)
plt.title("a1 vs a3")
plt.subplot(4,1,3)
plt.plot(a1,a4)
plt.title("a1 vs a4")
plt.subplot(4,1,4)
plt.plot(a2,a4)
plt.title("a2 vs a4")
```

```python
plt.figure(figsize=(15,4))
a1=np.arange(1,11)
a2=a1**2
a3=a1**3
a4=a1*2
plt.subplot(2,2,1)
plt.plot(a1,a2)
plt.title("a1 vs a2")
plt.subplot(2,2,2)
plt.plot(a1,a3)
plt.title("a1 vs a3")
plt.subplot(2,2,3)
plt.plot(a1,a4)
plt.title("a1 vs a4")
plt.subplot(2,2,4)
plt.plot(a2,a4)
plt.title("a2 vs a4")
```

```python
np.percentile(df["tip"],[25,30,90,56])
```

```python
np.quantile(df["tip"],[0.25,0.30,0.90,0.56])
```

```python
q1=np.percentile(df["tip"],[25])
q1
```

```python
q3=np.percentile(df["tip"],[75])
q3
```

```python
iqr=q3-q1
iqr
```

```python
up=q3+1.5*iqr
up
```

```python
ll=q1-1.5*iqr
ll
```

```python
df[df["tip"]>up[0]]
```

```python
df[df["tip"]<ll[0]]
```

# handling
- drop
- caping

```python
df1=df[df["tip"]>up[0]]
df1
```

```python
df["tip"]=np.where(df["tip"]>up[0],up[0],np.where(df["tip"]<ll[0],ll[0],df["tip"]))
```
